```python
# So let's first split the data
X = df.drop('price',axis=1)
y = df['price']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)

len(X_train)
```

```python
# Now apart from all of this
# We want that our model should first get converted into numeric data
# After which we will be running a for loop to understand the correlation and omit the redundancy

# So let's start it
# First import the encoding model
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
```

```python
col_list = X_train.select_dtypes(include=['object']).columns
col_list
```

```python
X_train[col_list]=X_train[col_list].apply(l.fit_transform)
```

```python
# Correlation matrix
corr = X_train.corr()
fig,ax = plt.subplots(figsize = (20,18))
sns.heatmap(corr,annot=True,ax=ax,cmap=plt.cm.Blues);
```

```python
# Independent correlations ( which can be skipped )

def independentCorr(data,threshold):
    cor_col = set()
    cor_data = data.corr()
    for i in range(len(cor_data.columns)):
        for j in range(i):
            if (cor_data.iloc[i,j])>threshold:
                colname = cor_data.columns[i]
                cor_col.add(colname)
    return cor_col
    
            
```

```python
corr_features = independentCorr(X_train,0.85)
len(set(corr_features))
```

```python
corr_features
```

```python
X_test.drop(corr_features,axis=1)
X_train.drop(corr_features,axis=1)
```

#### Lasso Model

```python
from sklearn import linear_model
m1 = linear_model.Lasso()
m1.fit(X_train,y_train)
```

```python
X_test[col_list]=X_test[col_list].apply(l.fit_transform)
```

```python
### Trying Ensemble regressor
from sklearn.ensemble import RandomForestRegressor
m2 = RandomForestRegressor(max_depth = 100, random_state = 2)
m2.fit(X_train,y_train)
```

```python
m2.get_params()
```

```python
m2.score(X_test,y_test)
```

```python
y_preds = m1.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_preds,y_test)
```

```python
## HyperTuning parameters
## To use GridSearchCV

grid_2={
    
    'n_estimators': [10, 100, 500],
     'max_depth': [None],
     'max_features': ['auto', 'sqrt'],
     'min_samples_split': [6],
     'min_samples_leaf': [1, 2]
}

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

gscv = RandomizedSearchCV(estimator=m2,
                          param_distributions=grid_2,
                          # No need of this as brute force
                          cv=5,
                          verbose=2
                            )
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
m2 = RandomForestRegressor(n_estimators=100,
 min_samples_split= 6,
 min_samples_leaf= 1,
 max_features = 'auto',
 max_depth =None)
m2.fit(X_train,y_train)
```

```python
z = m2.score(X_test,y_test)
```

```python
## After Hypertuning this is the best accuracy that we get
print("Accuracy of our model: ",z*100,"%")
```
